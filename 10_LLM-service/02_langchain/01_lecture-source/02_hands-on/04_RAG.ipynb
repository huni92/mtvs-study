3. Document ìƒì„±

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("./snow-white.pdf")

docs = loader.load()
docs
```

4. split -> embed -> vector storeì— ì €ì¥

```python
from langchain_chroma import Chroma
from langchain.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter

# ì„ë² ë”©
embedding_model = OllamaEmbeddings(model="gemma2:latest")

# ë¬¸ì„œ ë¶„í• 
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=10)
texts = text_splitter.split_documents(docs)

# ë¶„í•œë  í…ìŠ¤íŠ¸ë¥¼ ì„ë² ë”© í•œ í›„ ë²¡í„°ìŠ¤í† ì–´ì— ì €ì¥
vertorstore = Chroma.from_documents(
    texts,
    embedding=embedding_model
)

vertorstore
```

5. VectorStoreì—ì„œ ìœ ì‚¬ë„ê°€ ë†’ì€ ë¬¸ì¥ ê²€ìƒ‰

```python
sample = "ì™•ë¹„ì™€ ë°±ì„¤ê³µì£¼ ì¤‘ ëˆ„ê°€ ë” ì•„ë¦„ë‹¤ìš¸ê¹Œ?"

# vertorstore.similarity_search_with_score(sample)

retriever = vertorstore.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 5}
)

retriever.batch([sample])
```

6. ìœ ì‚¬ë„ ë†’ì€ ë¬¸ì¥ê³¼ í•¨ê»˜ í”„ë¡¬í”„íŠ¸ ìƒì„±

```python
from langchain_core.prompts import ChatPromptTemplate

message = """
Answer this question using the provied context only.

{question}

Context:
{context}
"""

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "ë‹¹ì‹ ì€ ì–´ë¦° ì•„ì´ì—ê²Œ ê¿ˆê³¼ í¬ë§ì„ ì‹¬ì–´ì£¼ëŠ” ìœ ì¹˜ì› êµì‚¬ì…ë‹ˆë‹¤. ì§ˆë¬¸í•˜ëŠ” ì•„ì´ì˜ ë°˜ì‘ì— ìµœëŒ€í•œ í˜¸ì‘í•´ì£¼ë©° context ê¸°ë°˜ìœ¼ë¡œë§Œ ë‹µë³€í•´ì£¼ì„¸ìš”."),
        ("human", message)
    ]
)

question = "ë°±ì„¤ê³µì£¼ì˜ ì´ë¦„ì€ ì™œ ë°±ì„¤ê³µì£¼ì•¼?"

prompt.invoke({"context": retriever, "question": question})
```

7. chain ìƒì„± í›„ LLM ì¶”ë¡ 

```python
from langchain.chat_models import ChatOllama
from langchain_core.runnables import RunnablePassthrough

model = ChatOllama(
    model="gemma2:2b",
    temperature=0.3
)

chain = {"context": retriever, "question": RunnablePassthrough()} | prompt | model

chain.invoke("ë°±ì„¤ê³µì£¼ì™€ ì™•ë¹„ ì¤‘ ëˆ„ê°€ ë” ì•„ë¦„ë‹¤ìš¸ê¹Œ?")
```
